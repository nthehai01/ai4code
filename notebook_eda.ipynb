```python
from data import Dataset
import matplotlib.pyplot as plt
import numpy as np
```

```python
d = Dataset("data", "bert-base-cased", 100000, 100000, -1, 100000, 10)
df = d.load_dataset()
```

```python
# num cells

cell = df.groupby("id").count()

plt.hist(cell["cell_id"])
plt.show()
```

```python
# max len

df["cell_len"] = 0
df["word_count"] = df["source"].str.split().str.len()

plt.hist(df["cell_len"])
plt.show()
```

```python
np.mean(cell["cell_id"] <= 128)  # 0.96
```
